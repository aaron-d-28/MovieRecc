```python
full_movie = movies.merge(credit, on = 'title')
full_movie.shape
```

```python
# genreabs
# id
# keywords 
# original language 
#  title 
# overview 
# cast 
# / crew 
```

```python
movies = full_movie[['movie_id','title', 'overview' , 'cast' , 'crew' , 'genres' , 'keywords']]
movies.count()
```

```python
movies.dropna(inplace=True)
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def converttogenre(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def converttodirector(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
import ast

def converttocast(obj):
    L = []  
    try:
        newobj = ast.literal_eval(obj)  # Convert string to list of dicts

        # Extract names safely (handles lists with < 3 elements)
        for i in range(min(3, len(newobj))):
            L.append(newobj[i].get('name', 'Unknown'))  # Default to 'Unknown' if key is missing

    except (ValueError, SyntaxError):  # If the string isn't valid JSON-like
        print("Invalid input format!")
        return []
    
    return L
```

```python
movies.cast = movies.cast.apply(converttocast)
```

```python
movies.head(1)
```

```python
movies.crew = movies.crew.apply(converttodirector)
```

```python
movies.head(1)
```

```python
movies.overview = movies['overview'].fillna('').apply(lambda x : x.split())
```

```python
movies.genres = movies.genres.apply(converttogenre)
movies.keywords = movies.keywords.apply(converttogenre)
```

```python
movies.head()
```

```python
movies.genres = movies.genres.apply(lambda x:[i.replace(" ","") for i in x])
movies.keywords = movies.keywords.apply(lambda x:[i.replace(" ","") for i in x])
movies.cast = movies.cast.apply(lambda x:[i.replace(" ","") for i in x])
movies.crew = movies.crew.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] =  movies.cast + movies.crew + movies.genres +movies.overview + movies.keywords
```

```python
movies.tags.head()
```

```python
new_movies= movies[['movie_id','title','tags']]
```

```python
new_movies.head()
```

```python
new_movies.tags= new_movies.tags.apply(lambda x:" ".join(x))
```

```python
new_movies.tags = new_movies.tags.apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
new_movies.tags = new_movies.tags.apply(stem)
```

```python
vectors = cv.fit_transform(new_movies.tags).toarray()
vectors
```

```python
words = cv.get_feature_names_out()
words[737]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(txt):
    y=[]

    for i in txt.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity[0]
```

```python
def reccomend_movies_json(moviename):
    movie_index = new_movies[new_movies.title==moviename].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    movie_reccomend_index=[index for index,values in movies_list]
    movie_name = new_movies.iloc[movie_reccomend_index]
    moviesval = movie_name[['movie_id','title']]
    movie_jsonified = moviesval.to_json(orient="records")
    return json.loads(movie_jsonified)
```

```python
movies = reccomend_movies_json('Batman Begins')
movies
```

```python
import pickle 
pickle.dump(reccomend_movies_json,open('movie_reccomendation_func.pkl','wb'))
```
